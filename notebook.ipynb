# Player data

Each player has the following clusters of data:

$Basic$ $Data$ - id, name, full name, team, team code, position

$Player$ $Data$ $For$ $The$ $Coming$ $Round$ - chance of playing this round, chance of paying next round, form

$Player$ $Dynamic$ $Data$ - selected by % of managers, transers in/out total, transfers in/out for the coming gameweek, news and the date for the new

$Player$ $Season$ $Data$ - dream team count, total points, total minutes

$Prolific$ $Data$ - goals scored, assists, penalties missed

$Defense$ $Data$ - saves, goals conceded, clean sheets, penalties saved, own goals

$Discipline$ $Data$ - yellow cards, red cards

$Season$ $Numerical$ $Data$ - bps, bonus, influence, creativity, threat, ict_index, influence_rank, creativity_rank, threat_rank, ict_index_rank, ict_index_rank_type, creativity_rank_type, threat_rank_type, influence_rank_type

$Points$ $For$ $The$ $Current$ $Gameweek$ - points for the current gameweek

$Price$ $Data$ - current price, price changed before the coming gameweek, price change for the season, price drop before the coming gameweek, price drop for the season

$Set$ $Piece$ - corners order, freekicks order, penalties order

$Misc$ - is player owned by the user

# Decode

Let us decode the team code to the team name

```python
from objects.team import *

past_fixture_dict = {

}

for fixture in fixtures_data['past_fixtures']:
    if fixture['kickoff_time'] == None:
        continue
    team_a = fixture['team_a']
    team_h = fixture['team_h']

    team_a_name = get_team_short_name(get_team_by_id(team_a))
    team_h_name = get_team_short_name(get_team_by_id(team_h))

    fixture['team_a'] = team_a_name
    fixture['team_h'] = team_h_name
    past_fixture_dict[fixture['id']] = fixture
```

# Decode Stats

```python
# sort past fixture by kickoff time
all_fixtures_dict = {k: v for k, v in sorted(past_fixture_dict.items(), key=lambda item: item[1]['kickoff_time'])}

all_fixtures_df = pd.DataFrame(all_fixtures_dict)
all_fixtures_df
```

```python
# get the past fixtures the ones where finished is True
finished_games = {}
upcoming_games = {}

for val, fixt in enumerate(past_fixture_dict):
    if (past_fixture_dict[fixt]['finished']):
        finished_games[val] = past_fixture_dict[fixt]
    else:
        upcoming_games[val] = past_fixture_dict[fixt]

# get dataframe of finished games
finished_games_df = pd.DataFrame(finished_games)
finished_games_df

```

```python
# List of games that haven't happened yet during this season

upcoming_games_df = pd.DataFrame(upcoming_games)
upcoming_games_df
```

```python
from fantasy_api.get_fixtures import get_fixtures_for_specific_week

this_gameweek = 19
fixtures_for_this_gameweek = get_fixtures_for_specific_week(this_gameweek)

a = 0 
for fixture in fixtures_for_this_gameweek:
    team_a = fixture['team_a']
    team_h = fixture['team_h']

    team_a_name = get_team_short_name(get_team_by_id(team_a))
    team_h_name = get_team_short_name(get_team_by_id(team_h))

    fixture['team_a'] = team_a_name
    fixture['team_h'] = team_h_name
    fixtures_for_this_gameweek[a] = fixture
    a += 1

# get the dataframe of fixtures for this gameweek
fixtures_for_this_gameweek_df = pd.DataFrame(fixtures_for_this_gameweek)
fixtures_for_this_gameweek_df
```

# Let us dig deep into the stats section of the fixtures

Each fixture contains a list of events that happened during the match and affected the point calculation.

These are the variables that we are interested in:

$goals$ $scored,$ $assists,$ $own$ $goals,$ $penalties$ $saved,$ $penalties$ $missed,$ $yellow$ $cards,$ $red$ $cards,$ $saves,$ $bonus,$ $bps$

Note: The values with two words follow the python convention of using underscores to separate words.

<br></br>
<br></br>
Each of these identifiers is attached to a list of a and h values. It contains the number of times an event happened and the id of the player that performed the event.

They are clustered into two groups (a) for away and (h) for home. This is an example of how the data is stored in the fixture stats.

# Data Summary

Now that we have a feel for the data, let us summarize our inputs and define desired outputs. 

We have a list of players, each of which has a set of data that is labeled as belonging to a specific cluster. No data can belong to more than one cluster. These clusters are defined in the notebook.

On top of the list of players, we have a summary of team data for each team. The data is very succinct, but it tells us a lot about the team.

**strength_overall_home, strength_overall_away,	strength_attack_home,	strength_attack_away,	strength_defence_home,	strength_defence_away**

Let us compute the properties of each data point and understand the variability in the data.

```python
import numpy as np
import matplotlib.pyplot as plt

strength_overall_home_mean = np.mean(clubs_df['strength_overall_home'])
strength_overall_home_std = np.std(clubs_df['strength_overall_home'])
strength_overall_home_max = np.max(clubs_df['strength_overall_home'])
strength_overall_home_median = np.median(clubs_df['strength_overall_home'])
strength_overall_home_min = np.min(clubs_df['strength_overall_home'])

sohm = [a['strength_overall_home'] for a in data['teams']]
short_names = [a['short_name'] for a in data['teams']]


X = np.arange(1000)
fig = plt.figure()
ax = fig.add_axes([0,1,2,2])
ax.set_yticks((1000,1400, 1000,1050,1100,1150,1200,1250,1300,1350,1400))

for val, team in enumerate(short_names):
    normal_prob = (round(sohm[val]/ (strength_overall_home_mean * 20), 4))

    ax.bar(team, sohm[val], color = 'b', width = 0.1)
    ax.text(team, sohm[val], str(normal_prob), color = 'black', fontsize = 12, fontname = 'Helvetica')


ax.bar("Mean", strength_overall_home_mean, color = 'r', width = 0.1)
ax.text("Mean", strength_overall_home_mean, str(strength_overall_home_mean), color = 'black', fontsize = 12, fontname='Helvetica')

ax.bar("Median", strength_overall_home_median, color = 'g', width = 0.1)
ax.text("Median", strength_overall_home_median, str(strength_overall_home_median), color = 'black', fontsize = 12, fontname='Helvetica')

ax.bar("Min", strength_overall_home_min, color = 'y', width = 0.1)
ax.text("Min", strength_overall_home_min, str(strength_overall_home_min), color = 'black', fontsize = 12, fontname='Helvetica')

ax.bar("Max", strength_overall_home_max, color = 'c', width = 0.1)
ax.text("Max", strength_overall_home_max, str(strength_overall_home_max), color = 'black', fontsize = 12, fontname='Helvetica')

ax.bar("Std", strength_overall_home_std, color = 'm', width = 0.1)
ax.text("Std", strength_overall_home_std, str(round(strength_overall_home_std)), color = 'black', fontsize = 12, fontname='Helvetica')

ax.set_title("Overall Strength of Premier League Teams When Playing Home")
ax.set_xlabel("Team", fontsize = 12, fontname = 'Helvetica')
ax.set_ylabel("Rating", fontsize = 12, fontname = 'Helvetica')


```
