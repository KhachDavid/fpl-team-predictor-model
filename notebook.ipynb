**Player data**

Each player has the following clusters of data:

$Basic$ $Data$ - id, name, full name, team, team code, position

$Player$ $Data$ $For$ $The$ $Coming$ $Round$ - chance of playing this round, chance of paying next round, form

$Player$ $Dynamic$ $Data$ - selected by % of managers, transers in/out total, transfers in/out for the coming gameweek, news and the date for the new

$Player$ $Season$ $Data$ - dream team count, total points, total minutes

$Prolific$ $Data$ - goals scored, assists, penalties missed

$Defense$ $Data$ - saves, goals conceded, clean sheets, penalties saved, own goals

$Discipline$ $Data$ - yellow cards, red cards

$Season$ $Numerical$ $Data$ - bps, bonus, influence, creativity, threat, ict_index, influence_rank, creativity_rank, threat_rank, ict_index_rank, ict_index_rank_type, creativity_rank_type, threat_rank_type, influence_rank_type

$Points$ $For$ $The$ $Current$ $Gameweek$ - points for the current gameweek

$Price$ $Data$ - current price, price changed before the coming gameweek, price change for the season, price drop before the coming gameweek, price drop for the season

$Set$ $Piece$ - corners order, freekicks order, penalties order

$Misc$ - is player owned by the user
